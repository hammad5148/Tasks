```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sks
```

```python
df = pd.read_csv('/content/icc-test-cricket-runs/ICC Test Batting Figures.csv', encoding='latin1',usecols=['Player',    'Span', 'Mat',  'Inn',  'NO',   'Runs', 'HS',   'Avg',  '100','50',     '0'],nrows=1000)
```

```python
df.head(5)
```

```python
df.shape
```

```python
df.info()
```

```python
sns.histplot(x='Avg',data=df)
sns.set_style('whitegrid')
plt.xticks(np.arange(0,100,5))
plt.xlabel('Average')
plt.ylabel('No. of Batter')
plt.title('Histogram of Test Average of 1000 Cricketer')
plt.show()
```

```python
sns.histplot(df['Runs'])
plt.xticks(np.arange(0,15000,1000),rotation=90)
plt.yticks(np.arange(0,350,25))
plt.xlabel('Runs')
plt.ylabel('No. of Batter')
```

```python
df[df['Runs']>9000]
```

```python
sns.scatterplot(x='Inn',y='100',data=df)
plt.xlabel('Innings Played')
plt.ylabel('Century scored')
```

```python
sns.scatterplot(x='Inn',y='Avg',data=df)
plt.xlabel('Innings Played')
plt.ylabel('Average')
plt.title('Average vs Innings played')
```

```python
sns.scatterplot(x='Inn',y='NO',data=df)
plt.xlabel('Innings Played')
plt.ylabel('Not Out')
plt.title('Not Outs vs Innings Played')
```
